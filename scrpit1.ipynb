```python
# CARREGAMENTO DAS BIBLIOTECAS PANDAS E MATPLOTLIB
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CRIANDO O OBJETO DF A PARTIR DA LEITURA DO DATAFRAME IMIGRANTES_CANADA
df = pd.read_csv('imigrantes_canada.csv')
```

```python
# VISUALIZANDO O DATAFRAME
df
```

```python
# UTILIZANDO O PARAMETRO .INFO() E POSSIVEL TER ACESSO A INFORMAÇÕES IMPORTANTES DO DATAFRAME 
df.info()
```

```python
# ESTE CODIGO DEFINE COMO INDICE DO DF A COLUNA PAIS
# O ARGUMENTO INPLACE = TRUE MODIFICA O DF DIRETAMENTE SEM CRIAR UMA COPIA
df.set_index('País', inplace=True)
```

```python
# ESTE CODIGO CRIA A LISTA CHAMADA ANOS CONTENDO STRINGS QUE REPRESENTAM OS ANOS DE 1980 A 2013
# LIST(...): TRANSFORMA O RESULTADO DO MAP EM UMA LISTA
# MAP(STR, ...): CONVERTE CADA NÚMERO DA SEQUÊNCIA EM UMA STRING
# RANGE(1980, 2014): GERA UMA SEQUÊNCIA DE NÚMEROS
anos = list(map(str, range(1980, 2014)))
```

```python
# ESTE BLOCO SELECIONA UMA PARTE ESPECIFICA DO DF COM BASE NO INDICE E NAS COLUNAS
# .LOC ACESSA OS DADOS NO DF
# 'Brasil' SE REFERE AO INDICE (LINHA) CUJO VALOR É 'Brasil'
brasil = df.loc['Brasil', anos]
```

```python
# E CRIADO UM NOVO DATAFRAME CHAMADO DADOS_BRASIL A PARTIR DE UM DICIONARIO
# BRASIL.INDEX.TOLIST() CONVERTE O INDICE DO OBJETO BRASIL EM UMA LISTA, NO CASO OS ANOS
# BRASIL.VALUES.TOLIST() CONVERTE OS VALORES CORRESPONDENTES EM UMA LISTA, NO CASO O NUMERO DE IMIGRANTES
# O DICIONARIO TERÁ DUAS CHAVES: ANO E IMIGRANTES, COM LISTAS DE ANOS E SEUS VALORES 
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
# VISUALIZANDO O DATAFRAME 
dados_brasil
```

```python
# CRIANDO O GRÁFICO 1 
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', 
            '1990', '1995', 
            '2000', '2005', '2010'])
plt.show()
```

```python
# CRIANDO O GRÁFICO 2 A PARTIR DE UMA FIGURA
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=2.5, color='g')
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
# ESTE CODIGO CRIA UM NOVO DATAFRAME COM OS DADOS DA AMERICA DO SUL
# CONTENDO APENAS AS LINHAS DO DATAGRAME DF ONDE A REGIÃO E "AMÉRICA DO SUL"
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
# ESTE CODIGO CRIA A PARTIR DO DF AMERICA_SUL UM NOVO DATAFRAME CONTENDO OS MESMOS DADOS
# MAS ORDENADOS DE FORMA QUE AS LINHAS COM OS VALORES MENORES NA COLUNA TOTAL APAREÇAM PRIMEIRO
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
# CRIANDO O GRÁFICO 3 EM BARRAS HORIZONTAIS
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')


fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('America do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no periodo de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), 
            color='black', fontsize=10, 
            ha='left', va='center')

 
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
# O SCRIPT CRIA UMA FIGURA COM QUATRO SUBPLOTS 

fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatro maiores países da America do Sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')


ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```
